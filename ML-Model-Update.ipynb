**Introducción**

En la era de la medicina digital y la revolución del big data, el uso de algoritmos de Machine Learning (ML) ha transformado el panorama de la atención médica, ofreciendo herramientas que pueden mejorar el análisis predictivo y el diagnóstico clínico. Estas técnicas de ML se han convertido en un componente esencial para predecir resultados de pacientes, optimizar la gestión hospitalaria y mejorar el diagnóstico y tratamiento de enfermedades..

**Objetivo general**

El objetivo principal de este proyecto es implementar un modelo basado en algoritmos de Machine Learning que mejore la capacidad de predicción y análisis de diagnósticos clínicos en la empresa "RedLab Perú S.A.C.". Esta implementación busca aumentar la precisión, reducir errores humanos y proporcionar un soporte más efectivo para los profesionales de la salud.

**Aplicación de la metodología CRISP-ML(Q)**

Para garantizar la efectividad y eficiencia del proyecto, hemos decidido hacer uso de la metodología CRISP-ML(Q), que nos guiará paso a paso en el desarrollo y despliegue del modelo. Las fases de este proceso son:

*   Comprensión del Negocio: En esta fase, identificamos y comprendemos las necesidades específicas y retos que enfrenta "RedLab Perú S.A.C." en el diagnóstico clínico, y cómo el Machine Learning puede ayudar a abordarlos.
*   Comprensión de los Datos: Recopilamos y examinamos los datos clínicos disponibles, comprendiendo su estructura, calidad y potenciales desafíos.
*   Preparación de los Datos: Limpiamos, transformamos y enriquecemos el conjunto de datos para asegurarnos de que esté listo para el modelado.
*   Modelado: Experimentamos con varios algoritmos de Machine Learning para encontrar el modelo que mejor se adapte a nuestras necesidades y proporcionar resultados óptimos.
*   Evaluación: Una vez construido el modelo, evaluamos su rendimiento y precisión, garantizando que cumpla con los estándares requeridos para un despliegue en un entorno clínico.
*   Despliegue: Finalmente, integramos el modelo en el sistema existente de "RedLab Perú S.A.C.", permitiendo a los profesionales de la salud hacer uso de esta herramienta avanzada en sus prácticas diarias.

**Relevancia**

```python
def extract_and_clean_column(df, column_name):
    df[column_name] = df[column_name].astype(str)
    return df[column_name].str.extract(r'([a-zA-Z]+)'), df[column_name].str.extract(r'([\d.]+)')

def process_dataframe(df):
    df['Result Label'], df['Result'] = extract_and_clean_column(df, 'Result')
    col_order = ['Result', 'Result Label'] + [col for col in df if col not in ['Result', 'Result Label']]
    return df[col_order].drop(columns=df.filter(regex='^\d+$').columns)

def safe_to_datetime(series):
    return pd.to_datetime(series.replace(r":(\d{2})\d+", r":\1", regex=True), errors='coerce')

# Suponiendo que df_base es tu DataFrame
df_base = process_dataframe(df_base)
df_base['Application time'] = safe_to_datetime(df_base['Application time'])
df_base['Complete Time'] = safe_to_datetime(df_base['Complete Time'])
```

```python
def process_time(df):
    # Asegurar el formato de fecha y hora para las columnas relevantes
    df['Application time'] = pd.to_datetime(df['Application time'])
    df['Complete Time'] = pd.to_datetime(df['Complete Time'])

    # Calcular la duración
    df['Duration'] = df['Complete Time'] - df['Application time']

    # Extraer la hora de inicio y categorizar el periodo del día
    hours = df['Application time'].dt.hour
    conditions = [
        (6 <= hours) & (hours < 12),
        (12 <= hours) & (hours < 17),
        (17 <= hours) & (hours < 21),
    ]
    choices = ['Morning', 'Afternoon', 'Evening']
    df['Period of Day'] = np.select(conditions, choices, default='Night')

    # Seleccionar las columnas relevantes para la reorganización
    cols = ['Duration', 'Period of Day'] + [c for c in df.columns if c not in ['Duration', 'Period of Day']]
    return df[cols]

# Procesar el DataFrame df_base
df_base = process_time(df_base)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt

# Preprocesamiento y limpieza de datos
def preprocess_data(df_patient, df_train_gender):
    df_patient['NOMBRE Y APELLIDOS'] = df_patient['NOMBRE Y APELLIDOS'].str.lower().str.strip()
    df_patient.dropna(subset=['NOMBRE Y APELLIDOS'], inplace=True)
    df_train_gender.columns = ['Name', 'Gender', 'Precision']
    df_train_gender['Name'] = df_train_gender['Name'].str.lower().str.strip()

    return df_patient, df_train_gender

df_patient, df_train_gender = preprocess_data(df_patient, df_train_gender)

# Construcción y evaluación del modelo
def build_and_evaluate_model(df_train_gender):
    X_train, X_test, y_train, y_test = train_test_split(
        df_train_gender['Name'], df_train_gender['Gender'],
        test_size=0.2, random_state=42
    )

    vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(2, 3))
    clf = LogisticRegression(random_state=42, max_iter=1000)
    clf.fit(vectorizer.fit_transform(X_train), y_train)
    y_pred = clf.predict(vectorizer.transform(X_test))

    # Métricas
    scores = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, pos_label='M'),
        'Recall': recall_score(y_test, y_pred, pos_label='M'),
        'F1 Score': f1_score(y_test, y_pred, pos_label='M')
    }

    return clf, vectorizer, scores, y_test, y_pred

clf, vectorizer, scores, y_test, y_pred = build_and_evaluate_model(df_train_gender)

# Visualización de métricas y matriz de confusión
def plot_metrics_and_confusion_matrix(scores, y_test, y_pred):
    sns.set_style("whitegrid")
    fig, axs = plt.subplots(1, 2, figsize=(18, 6))

    # Gráfico de métricas
    axs[0].bar(scores.keys(), scores.values(), color=sns.color_palette("coolwarm", len(scores)), edgecolor="black")
    axs[0].set(title="Performance Metrics", xlabel="Metric", ylabel="Value")
    axs[0].set_ylim(0, 1.1)

    # Matriz de confusión
    conf_matrix = confusion_matrix(y_test, y_pred)
    sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues', ax=axs[1],
                xticklabels=['Predicted M', 'Predicted F'],
                yticklabels=['Actual M', 'Actual F'])
    axs[1].set(title='Confusion Matrix', xlabel='Predicted Labels', ylabel='True Labels')

    plt.tight_layout()
    plt.show()

plot_metrics_and_confusion_matrix(scores, y_test, y_pred)

# Función para predecir el género
def get_gender(clf, vectorizer, name):
    pred_gender = clf.predict(vectorizer.transform([name]))[0]
    prob = clf.predict_proba(vectorizer.transform([name]))[0]
    return "M" if prob[0] > prob[1] else "F"

# Aplicación del modelo al DataFrame
def apply_model_to_dataframe(df_patient, clf, vectorizer):
    df_patient['GENDER_ML'] = df_patient['NOMBRE Y APELLIDOS'].str.split().apply(lambda x: get_gender(clf, vectorizer, x[-1]))
    return df_patient

df_patient = apply_model_to_dataframe(df_patient, clf, vectorizer)
```

```python
# Esta función convierte los nombres de una columna del dataframe en minúsculas y elimina espacios adicionales.
def normalize_names(df, column_name):
    df[column_name] = df[column_name].str.lower().str.strip()

# Esta función combina dos dataframes basados en nombres de columnas proporcionados.
def merge_dfs_on_name(df1, df2, df1_col, df2_col):
    return df1.merge(df2, left_on=df1_col, right_on=df2_col, how='left')

# Esta función reordena las columnas de un dataframe, insertando las nuevas columnas después de una columna específica.
def reorder_columns(df, base_columns, insert_after_col, new_columns):
    desired_order = base_columns.copy()
    name_index = desired_order.index(insert_after_col)

    for col in new_columns:
        desired_order.insert(name_index + 1, col)
        name_index += 1

    return df[desired_order]

# Esta función determina el rango de tiempo basado en dos columnas de tiempo.
def determine_time_range(row, ranges):
    start_hour = row['Application time'].hour
    end_hour = row['Complete Time'].hour

    for r in ranges:
        if start_hour >= r['start'] and end_hour <= r['end']:
            return r['label']
    return 'Out of Range'

# Esta es la función principal que ejecuta las operaciones anteriores en orden.
def main():
    global df_final  # Define df_final como una variable global para poder acceder fuera de esta función.

    # Normaliza los nombres de las columnas 'Name' y 'NOMBRE Y APELLIDOS' en ambos dataframes.
    normalize_names(df_base, 'Name')
    normalize_names(df_patient, 'NOMBRE Y APELLIDOS')

    # Combina los dos dataframes basados en los nombres.
    df_merged = merge_dfs_on_name(df_base, df_patient, 'Name', 'NOMBRE Y APELLIDOS')

    # Reordena las columnas y renombra algunas.
    df_final = reorder_columns(df_merged, df_base.columns.tolist(), 'Name', ['EDAD'])
    df_final = df_final.rename(columns={'EDAD': 'age', 'Assay': 'assay', 'Result': 'result'})

    # Convierte las columnas 'Application time' y 'Complete Time' a objetos datetime para poder trabajar con fechas y horas.
    df_final['Application time'] = pd.to_datetime(df_final['Application time'])
    df_final['Complete Time'] = pd.to_datetime(df_final['Complete Time'])

    # Define los rangos de tiempo.
    time_ranges = [
        {"start": 7, "end": 9, "label": "7 - 9 a.m."},
        {"start": 8, "end": 12, "label": "8 - 12 a.m."},
        {"start": 12, "end": 17, "label": "12 - 5 p.m."},
        {"start": 15, "end": 19, "label": "3 - 7 p.m."}
    ]

    # Aplica la función determine_time_range para determinar el rango de tiempo para cada fila.
    time_range_series = df_final.apply(determine_time_range, ranges=time_ranges, axis=1)

    # Inserta la nueva columna 'Time Range' después de 'Complete Time'.
    col_idx = df_final.columns.get_loc('Complete Time')
    df_final.insert(col_idx + 1, 'time_range', time_range_series)

    # Guarda el dataframe final en un archivo CSV.
    #df_final.to_csv('df_final_prueba.csv', index=False)

# Ejecuta la función main.
main()
```

```python
def create_patient_labels(df):
    """
    Create a new column 'Patient_Label' based on unique names in the 'Name' column.
    The format of the labels will be 'Paciente n' where 'n' is a unique identifier.

    Parameters:
    - df (pd.DataFrame): Input DataFrame containing a 'Name' column.

    Returns:
    - pd.DataFrame: DataFrame with the added 'Patient_Label' column.
    """

    # Create unique numerical identifiers for each name
    df['Patient_ID'] = pd.factorize(df['Name'])[0] + 1

    # Convert these identifiers to the desired format
    df['Patient_Label'] = 'Paciente-' + df['Patient_ID'].astype(str)

    # Reorder the columns to place 'Patient_Label' right after 'Name'
    name_idx = df.columns.get_loc('Name')
    df.insert(name_idx + 1, 'Patient_Label', df.pop('Patient_Label'))

    # Drop the numerical 'Patient_ID' column and 'Name' column as they're not needed anymore
    df.drop(columns=['Patient_ID', 'Name'], inplace=True)

    return df

df_final = create_patient_labels(expanded_df)
df_final
```

```python
#JSON
# Crear un diccionario a partir de un DataFrame previamente definido llamado df_config_dict
sheets_dict = df_config_dict

# Funciones auxiliares
# Esta función recupera las configuraciones específicas para una prueba de ensayo dada.
def get_config_data(assay):
    # Filtrar las configuraciones que coinciden con el ensayo dado
    config_data = sheets_dict['config'][sheets_dict['config']['assay'] == assay]
    if config_data.empty:
        return []
    tipo_config_id_values = config_data.iloc[0]['tipo_config_id']
    return tipo_config_id_values.split(",") if isinstance(tipo_config_id_values, str) else [tipo_config_id_values]

# Esta función recupera las configuraciones que coinciden con un tipo_config_id y una fila específicos.
def get_matching_configs(tipo_config_id, row):
    tipo_config_rows = sheets_dict['tipo_config'][sheets_dict['tipo_config']['id'] == float(tipo_config_id)]
    for _, tipo_config_row in tipo_config_rows.iterrows():
        expected_entries = tipo_config_row['name'].split(",")
        if any(entry in row.to_string() for entry in expected_entries):
            config_sheet_name = tipo_config_row['dataframe_name']
            filtered_configs = sheets_dict[config_sheet_name][sheets_dict[config_sheet_name]['assay'] == row['assay']]
            # Filtrar por género y rango de tiempo si están disponibles en las configuraciones y en la fila
            if 'gender' in filtered_configs.columns and 'gender' in row:
                filtered_configs = filtered_configs[filtered_configs['gender'] == row['gender']]
            if 'time_range' in filtered_configs.columns and 'time_range' in row:
                filtered_configs = filtered_configs[filtered_configs['time_range'] == row['time_range']]
            return filtered_configs
    return pd.DataFrame()

# Esta función recupera las entradas utilizadas para un tipo_config_id específico.
def get_used_entries(tipo_config_id):
    tipo_config_rows = sheets_dict['tipo_config'][sheets_dict['tipo_config']['id'] == float(tipo_config_id)]
    if not tipo_config_rows.empty:
        return tipo_config_rows.iloc[0]['name']
    return None

# Esta función procesa una fila específica del DataFrame y devuelve un resultado de diagnóstico y paciente.
def revised_process_row_with_used_entries(row):
    diagnostic_result = None
    patient_result = None
    used_entries_set = set()

    for tipo_config_id in get_config_data(row['assay']):
        entries = get_used_entries(tipo_config_id)
        if entries:
            used_entries_set.update(entries.split(", "))

        for _, config in get_matching_configs(tipo_config_id, row).iterrows():
            # Intentar coincidir con las condiciones de configuración
            try:
                min_value = config['min_value'] if not pd.isna(config.get('min_value')) else float('-inf')
                max_value = config['max_value'] if not pd.isna(config.get('max_value')) else float('inf')
                result_value = float(row['result'])

                if 'age_min' in config and 'age_max' in config and not pd.isna(config['age_min']) and not pd.isna(config['age_max']):
                    if 'age' in row and not pd.isna(row['age']):
                        age_min = config['age_min']
                        age_max = config['age_max']
                        patient_age = float(row['age'])
                        if min_value <= result_value <= max_value and age_min <= patient_age <= age_max:
                            diagnostic_result = config['display_result_output']
                            patient_result = config['diagnostic_dict_output']
                            break
                else:
                    if min_value <= result_value <= max_value:
                        diagnostic_result = config['display_result_output']
                        patient_result = config['diagnostic_dict_output']
                        break
            except:
                continue
        if diagnostic_result and patient_result:
            break

    # Combinar diagnostic_result y patient_result en un único formato JSON
    combined_result = {
        "diagnostic_result": diagnostic_result,
        "patient_result": patient_result
    }

    return {"Final_Result": combined_result, "used_entries": ", ".join(used_entries_set)}

# Aplicar la configuración dinámica a todo el DataFrame y agregar columnas de resultados
def apply_dynamic_configurations_with_used_entries(df):
    df_results = df.apply(revised_process_row_with_used_entries, axis=1)
    df['Final_Result'] = df_results.apply(lambda x: x['Final_Result'])
    df['Used_Entries'] = df_results.apply(lambda x: x['used_entries'])

    col_order = df.columns.tolist()
    result_idx = col_order.index('result')
    for col in ['Final_Result', 'Used_Entries']:
        col_order.insert(result_idx + 1, col_order.pop(col_order.index(col)))
    return df[col_order]

# Esta función verifica si una fila específica coincide con una configuración dada.
def check_row_against_config(row, config):
    result_value = float(row['result'])  # Asegurarse de que result_value es un float
    for _, conf in config.iterrows():
        if 'min_value' in conf and 'max_value' in conf:
            if not (conf['min_value'] <= result_value <= conf['max_value']):
                continue

        if 'gender' in conf and not pd.isna(conf['gender']):
            if conf['gender'] != row['gender']:
                continue

        if 'age_min' in conf and 'age_max' in conf:
            if not (conf['age_min'] <= row['age'] <= conf['age_max']):
                continue

        return 1  # Si todas las condiciones se cumplen, regresar 1

    return 0  # Si ninguna configuración coincide, regresar 0

# Función para agregar las columnas basadas en las reglas del archivo de configuración a la fila del DataFrame
def add_config_columns(row):
    for sheet_name, config in sheets_dict.items():
        if sheet_name not in ['config', 'tipo_config']:
            row[sheet_name] = check_row_against_config(row, config)
    return row

# Aplicación de las funciones
df_final = df_final.apply(add_config_columns, axis=1)
df_final = apply_dynamic_configurations_with_used_entries(df_final)
```

```python
def expand_dataframe(df, target_size):
    """
    Expands the DataFrame by replicating existing rows to reach the target size.

    :param df: Input DataFrame.
    :param target_size: Target size of the DataFrame.
    :return: Expanded DataFrame.
    """
    repeats = target_size // len(df)
    extra_rows = target_size % len(df)
    return pd.concat([df] * repeats + [df.iloc[:extra_rows]]).reset_index(drop=True)

# Usage
df_final = expand_dataframe(df_final, 80000)
```

```python
# Logistic Regression
# Importando las bibliotecas necesarias
import pandas as pd
import json
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Función para intentar decodificar un JSON, o simplemente devolver un diccionario si ya lo es
def safe_json_or_dict(s):
    if isinstance(s, dict):
        return s
    try:
        return json.loads(s.replace("'", '"'))
    except Exception as e:
        print(f"Failed to decode: {s}. Error: {e}")
        return {}

# Convertir la columna 'Final_Result' en un diccionario
df_final['Final_Result_Dict'] = df_final['Final_Result'].apply(safe_json_or_dict)
df_final['Diagnostic_Result'] = df_final['Final_Result_Dict'].apply(lambda x: x.get('diagnostic_result', ''))

# Codificar los resultados diagnósticos
label_encoder = LabelEncoder()
df_final['Diagnostic_Result_Coded'] = label_encoder.fit_transform(df_final['Diagnostic_Result'].astype(str))

# Codificación one-hot para las columnas relevantes
columns_to_encode = ['time_range', 'gender', 'assay']
encoded_features = pd.get_dummies(df_final[columns_to_encode])

# Estandarizar las columnas 'age' y 'result'
age_scaler = StandardScaler().fit(df_final[['age']])
result_scaler = StandardScaler().fit(df_final[['result']])
scaled_age = age_scaler.transform(df_final[['age']])
scaled_result = result_scaler.transform(df_final[['result']])

# Crear conjunto de datos final
X = pd.concat([encoded_features, pd.DataFrame(scaled_age, columns=['age']), pd.DataFrame(scaled_result, columns=['result'])], axis=1)
y = df_final['Diagnostic_Result_Coded']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar el modelo de Regresión Logística
log_reg = LogisticRegression(max_iter=10000, random_state=42)
log_reg.fit(X_train, y_train)

# Hacer predicciones
y_pred = log_reg.predict(X_test)

# Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')

# Mostrar métricas
print(f"Accuracy: {accuracy}")
print(f"Recall: {recall}")
print(f"Precision: {precision}")

# Visualización de métricas
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 7))
ax = sns.barplot(x=['Accuracy', 'Recall', 'Precision'], y=[accuracy, recall, precision], palette="Blues_d")
plt.ylim(0, 1)
plt.title('Model Metrics Overview')
plt.ylabel('Score')
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    ax.text(p.get_x() + width/2., height + 0.02, '{:1.2f}'.format(height), ha="center")
plt.show()

# Función para hacer predicciones usando el modelo
def updated_predict_diagnosis_lr(**kwargs):
    data = pd.DataFrame([0] * len(X_train.columns), index=X_train.columns).transpose()
    for key, value in kwargs.items():
        if key + "_" + str(value) in data.columns:
            data[key + "_" + str(value)] = 1
        elif key in data.columns:
            data[key] = value
    if 'age' in kwargs:
        data['age'] = age_scaler.transform([[kwargs['age']]])[0][0]
    if 'result' in kwargs:
        data['result'] = result_scaler.transform([[kwargs['result']]])[0][0]

    predicted_code = log_reg.predict(data)
    predicted_label = label_encoder.inverse_transform(predicted_code)
    return {
        'diagnostic_dict_output': predicted_label[0]
    }
```

```python
# Gradient Boosting

# Importando las bibliotecas necesarias
import pandas as pd
import json
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Función para intentar decodificar un JSON, o simplemente devolver un diccionario si ya lo es
def safe_json_or_dict(s):
    if isinstance(s, dict):
        return s
    try:
        return json.loads(s.replace("'", '"'))
    except Exception as e:
        print(f"Failed to decode: {s}. Error: {e}")
        return {}

# Convertir la columna 'Final_Result' en un diccionario
df_final['Final_Result_Dict'] = df_final['Final_Result'].apply(safe_json_or_dict)
df_final['Diagnostic_Result'] = df_final['Final_Result_Dict'].apply(lambda x: x.get('diagnostic_result', ''))

# Codificar los resultados diagnósticos
label_encoder = LabelEncoder()
df_final['Diagnostic_Result_Coded'] = label_encoder.fit_transform(df_final['Diagnostic_Result'].astype(str))

# Codificación one-hot para las columnas relevantes
columns_to_encode = ['time_range', 'gender', 'assay']
encoded_features = pd.get_dummies(df_final[columns_to_encode])

# Estandarizar las columnas 'age' y 'result'
age_scaler = StandardScaler().fit(df_final[['age']])
result_scaler = StandardScaler().fit(df_final[['result']])
scaled_age = age_scaler.transform(df_final[['age']])
scaled_result = result_scaler.transform(df_final[['result']])

# Crear conjunto de datos final
X = pd.concat([encoded_features, pd.DataFrame(scaled_age, columns=['age']), pd.DataFrame(scaled_result, columns=['result'])], axis=1)
y = df_final['Diagnostic_Result_Coded']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar el modelo de Gradient Boosting
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_train)

# Hacer predicciones
y_pred = gb_model.predict(X_test)

# Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')

# Mostrar métricas
print(f"Accuracy: {accuracy}")
print(f"Recall: {recall}")
print(f"Precision: {precision}")

# Visualización de métricas
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 7))
ax = sns.barplot(x=['Accuracy', 'Recall', 'Precision'], y=[accuracy, recall, precision], palette="Blues_d")
plt.ylim(0, 1)
plt.title('Model Metrics Overview')
plt.ylabel('Score')
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    ax.text(p.get_x() + width/2., height + 0.02, '{:1.2f}'.format(height), ha="center")
plt.show()

# Función para hacer predicciones usando el modelo
def updated_predict_diagnosis_gb(**kwargs):
    data = pd.DataFrame([0] * len(X_train.columns), index=X_train.columns).transpose()
    for key, value in kwargs.items():
        if key + "_" + str(value) in data.columns:
            data[key + "_" + str(value)] = 1
        elif key in data.columns:
            data[key] = value
    if 'age' in kwargs:
        data['age'] = age_scaler.transform([[kwargs['age']]])[0][0]
    if 'result' in kwargs:
        data['result'] = result_scaler.transform([[kwargs['result']]])[0][0]

    predicted_code = gb_model.predict(data)
    predicted_label = label_encoder.inverse_transform(predicted_code)
    return {
        'diagnostic_dict_output': predicted_label[0]
    }
```

### **3.2. Elección de Modelo de Machine Learning**

**¿Qué es un árbol de decisión?**

Un árbol de decisión es una herramienta de modelado predictivo que mapea características (ramas) a conclusiones sobre el valor objetivo (hojas). Básicamente, divide un conjunto de datos en subconjuntos más pequeños y más homogéneos basándose en las características más significativas, haciendo esta tarea de manera recursiva de forma que las decisiones se toman en secuencia.

**Ventajas:**

- Interpretabilidad: Los árboles de decisión son modelos transparentes. Cada decisión se basa en una característica y un valor umbral, lo que hace que el proceso de decisión sea fácilmente rastreable y comprensible.

- Modelado de relaciones no lineales: A diferencia de los modelos lineales, los árboles de decisión pueden capturar y modelar relaciones no lineales en los datos, lo que los hace adecuados para conjuntos de datos con interacciones complejas.

- No necesita supuestos previos: Los árboles de decisión no hacen suposiciones fuertes sobre la distribución de los datos o la relación entre las características y la variable objetivo. Esto es diferente de los modelos lineales que suponen una relación lineal.

**Consideraciones:**

- Riesgo de sobreajuste: Si no se restringen adecuadamente, los árboles de decisión pueden crecer de manera descontrolada y memorizar los datos de entrenamiento, resultando en un sobreajuste. Es esencial aplicar técnicas de regularización como limitar la profundidad del árbol o requerir un número mínimo de puntos de datos en una hoja.

- Posibilidad de interpretación errónea: Aunque los árboles de decisión son a menudo elogiados por su interpretabilidad, si un árbol es muy grande o si se malinterpreta el significado de las divisiones, las decisiones del árbol pueden ser malinterpretadas.

```python
import pandas as pd
import numpy as np
import json
from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate, RepeatedStratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, make_scorer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.utils import class_weight
import ast

def safe_json_or_dict(s):
    try:
        return json.loads(s)
    except json.JSONDecodeError:
        try:
            return ast.literal_eval(s)  # Más seguro que eval()
        except (ValueError, SyntaxError) as e:
            print(f"Failed to decode or eval: {s}. Error: {e}")
            return {}

# Preprocesamiento de la columna 'Final_Result'
df_final['Final_Result'] = df_final['Final_Result'].astype(str)
df_final['Final_Result_Dict'] = df_final['Final_Result'].apply(safe_json_or_dict)
df_final['Diagnostic_Result'] = df_final['Final_Result_Dict'].apply(lambda x: x.get('diagnostic_result', ''))

# Codificar los resultados diagnósticos
label_encoder = LabelEncoder()
df_final['Diagnostic_Result_Coded'] = label_encoder.fit_transform(df_final['Diagnostic_Result'])

# Filtrar clases con una sola instancia
class_counts = df_final['Diagnostic_Result_Coded'].value_counts()
classes_to_keep = class_counts[class_counts > 1].index
df_final = df_final[df_final['Diagnostic_Result_Coded'].isin(classes_to_keep)]

# Reiniciar índices después de filtrar
df_final.reset_index(drop=True, inplace=True)

# Codificación one-hot para las columnas categóricas
columns_to_encode = ['time_range', 'gender', 'assay']
encoded_features = pd.get_dummies(df_final[columns_to_encode])

# Estandarizar las columnas numéricas
scaler = StandardScaler()
scaled_numerical = scaler.fit_transform(df_final[['age', 'result']])
scaled_numerical_df = pd.DataFrame(scaled_numerical, columns=['age', 'result'])

# Crear conjunto de datos final, asegurándote de que esté bien mezclado para evitar cualquier orden sesgado
X = pd.concat([encoded_features, scaled_numerical_df], axis=1).sample(frac=1, random_state=42)
y = df_final['Diagnostic_Result_Coded'].sample(frac=1, random_state=42)  # Asegúrate de usar el mismo random_state


# Verificar si todas las clases tienen al menos dos instancias para la estratificación
if (df_final['Diagnostic_Result_Coded'].value_counts() >= 2).all():
    stratify_param = y
else:
    stratify_param = None

# Dividir los datos en conjuntos de entrenamiento y prueba con estratificación (si es posible)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=stratify_param, random_state=42)

assert X.shape[0] == y.shape[0], "Las longitudes de X y y no coinciden."
# Asegúrate de que todas las clases tienen al menos dos instancias para la estratificación
y = df_final['Diagnostic_Result_Coded']
if (y.value_counts() >= 2).all():
    stratify_param = y
else:
    stratify_param = None

# Definir parámetros para GridSearchCV, ajustando el rango de ccp_alpha
params = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy'],
    'ccp_alpha': [0.001, 0.01, 0.1]  # Valores para la poda
}

# Instancia del modelo DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=42, class_weight='balanced')

# Instancia de RepeatedStratifiedKFold para la validación cruzada
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=42)
grid_search = GridSearchCV(estimator=dt_model, param_grid=params, cv=cv, n_jobs=-1, scoring='accuracy')

# Ajustar GridSearchCV con validación cruzada
grid_search.fit(X_train, y_train)

# Mejor modelo encontrado
best_dt_model = grid_search.best_estimator_

# Usar cross_validate para una evaluación más robusta con múltiples métricas
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'recall': make_scorer(recall_score, average='weighted'),
    'precision': make_scorer(precision_score, average='weighted'),
    'f1': make_scorer(f1_score, average='weighted')
}

scores = cross_validate(best_dt_model, X_train, y_train, cv=cv, scoring=scoring)

# Mostrar métricas promedio de la validación cruzada
print(f"Cross-Validated Accuracy: {np.mean(scores['test_accuracy'])}")
print(f"Cross-Validated Recall: {np.mean(scores['test_recall'])}")
print(f"Cross-Validated Precision: {np.mean(scores['test_precision'])}")
print(f"Cross-Validated F1 Score: {np.mean(scores['test_f1'])}")


def updated_predict_diagnosis_dt(model, label_encoder, age_scaler, result_scaler, **kwargs):
    data = pd.DataFrame([0] * len(X_train.columns), index=X_train.columns).transpose()
    for key, value in kwargs.items():
        if key + "_" + str(value) in data.columns:
            data[key + "_" + str(value)] = 1
        elif key in data.columns:
            data[key] = value
    if 'age' in kwargs:
        data['age'] = age_scaler.transform([[kwargs['age']]])[0][0]
    if 'result' in kwargs:
        data['result'] = result_scaler.transform([[kwargs['result']]])[0][0]

    predicted_code = model.predict(data)
    predicted_label = label_encoder.inverse_transform(predicted_code)
    return {
        'diagnostic_dict_output': predicted_label[0],
    }
```

**Elementos de la matriz de confusion**

```python
import numpy as np
from sklearn.metrics import confusion_matrix

# Asumimos que 'y_test', 'test_predictions' y 'X_test_with_assay' están definidos como se espera.
X_test_with_assay = X_test.join(df_final['assay'], how='left')

unique_assays = X_test_with_assay['assay'].unique()
metrics_per_assay = {}

# Esta función calcula los elementos de la matriz de confusión y devuelve los totales por clase.
def calculate_confusion_matrix_totals(conf_matrix):
    FP = conf_matrix.sum(axis=0) - np.diag(conf_matrix)
    FN = conf_matrix.sum(axis=1) - np.diag(conf_matrix)
    TP = np.diag(conf_matrix)
    TN = conf_matrix.sum() - (FP + FN + TP)
    return TP.sum(), TN.sum(), FP.sum(), FN.sum()

# Calcula las métricas agregadas para cada ensayo.
for assay in unique_assays:
    mask = X_test_with_assay['assay'] == assay
    y_true_assay = y_test[mask]
    y_pred_assay = test_predictions[mask]

    conf_matrix = confusion_matrix(y_true_assay, y_pred_assay)
    TP, TN, FP, FN = calculate_confusion_matrix_totals(conf_matrix)

    # Calcula las métricas globales para cada ensayo.
    accuracy = (TP + TN) / (TP + FP + FN + TN)
    precision = TP / (TP + FP) if TP + FP > 0 else 0
    recall = TP / (TP + FN) if TP + FN > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    # Almacena las métricas calculadas en el diccionario.
    metrics_per_assay[assay] = {
        'TP': TP,
        'TN': TN,
        'FP': FP,
        'FN': FN,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1_score
    }

# Imprime las métricas para cada ensayo.
for assay, metrics in metrics_per_assay.items():
    print(f"Metrics for assay '{assay}':")
    for metric_name, metric_value in metrics.items():
        if metric_name in ['Accuracy', 'Precision', 'Recall', 'F1 Score']:
            print(f"  {metric_name}: {metric_value:.2f}")
        else:
            print(f"  {metric_name}: {metric_value}")
    print("-" * 30)
```

```python
import pandas as pd
import joblib

# Carga los objetos necesarios para la predicción
def load_objects(model_file, objects_file):
    model = joblib.load(model_file)
    objects = joblib.load(objects_file)
    return model, objects

# Función para realizar una predicción con el modelo y los objetos cargados
def predict_with_loaded_objects(model, objects, input_data):
    # Crear un DataFrame con ceros para todas las columnas que el modelo espera
    data = pd.DataFrame([0] * len(objects['columns']), index=objects['columns']).T

    # Asegúrate de que tu entrada es un diccionario con todos los datos necesarios
    for key, value in input_data.items():
        if key in ['age', 'result']:  # Si la clave es 'age' o 'result', escalarlos
            scaled_value = objects[f'{key}_scaler'].transform([[value]])[0][0]
            data[key] = scaled_value
        else:
            # Para las variables categóricas que han sido codificadas en one-hot
            column_name = f"{key}_{value}"
            if column_name in data.columns:
                data[column_name] = 1

    # Realizar la predicción con el modelo
    prediction_code = model.predict(data)
    prediction_label = objects['label_encoder'].inverse_transform(prediction_code)

    return prediction_label[0]

# Usando las funciones definidas anteriormente para hacer una predicción
model, objects = load_objects('decision_tree_model_trained.pkl', 'training_objects.pkl')

# Suponiendo que esta es la nueva entrada para hacer la predicción
new_input_data = {
    'assay': 'VB12 Ⅱ',
    'result': 100
}

# Hacer la predicción
prediction = predict_with_loaded_objects(model, objects, new_input_data)
print(f"Predicted label: {prediction}")
```
